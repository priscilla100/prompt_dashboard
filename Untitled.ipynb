```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

class NL2AtomicAnalyzer:
    def __init__(self, df1, df2, experiment1_name="Dataset 1", experiment2_name="Dataset 2"):
        """
        Initialize analyzer with two datasets
        Expected columns: Model, Approach, Total_Samples, Accuracy, Precision, Recall, F1, Jaccard, Levenshtein
        """
        self.df1 = df1.copy()
        self.df2 = df2.copy()
        self.exp1_name = experiment1_name
        self.exp2_name = experiment2_name
        
        # Add experiment labels
        self.df1['Experiment'] = self.exp1_name
        self.df2['Experiment'] = self.exp2_name
        
        # Combine datasets
        self.df_combined = pd.concat([self.df1, self.df2], ignore_index=True)
        
        # Professional color palette
        self.colors = {
            'primary': '#2E86AB',
            'secondary': '#A23B72', 
            'accent1': '#F18F01',
            'accent2': '#C73E1D',
            'accent3': '#8B5A2B',
            'neutral': '#6C757D'
        }
        
        # Paper styling
        self.paper_style = {
            'font_family': 'Times New Roman',
            'font_size': 12,
            'title_font_size': 16,
            'legend_font_size': 10
        }
    
    def create_comprehensive_comparison_table(self):
        """Create a detailed comparison table"""
        # Get unique models and approaches
        models = sorted(set(self.df1['Model'].unique()) | set(self.df2['Model'].unique()))
        
        comparison_data = []
        
        for model in models:
            # Get data for this model from both datasets
            model_df1 = self.df1[self.df1['Model'] == model]
            model_df2 = self.df2[self.df2['Model'] == model]
            
            if len(model_df1) > 0 and len(model_df2) > 0:
                # Multiple approaches per model
                for _, row1 in model_df1.iterrows():
                    # Find matching approach in dataset 2
                    matching_row2 = model_df2[model_df2['Approach'] == row1['Approach']]
                    
                    if len(matching_row2) > 0:
                        row2 = matching_row2.iloc[0]
                        
                        comparison_data.append({
                            'Model': model,
                            'Approach': row1['Approach'],
                            f'Samples_{self.exp1_name}': int(row1['Total_Samples']),
                            f'Samples_{self.exp2_name}': int(row2['Total_Samples']),
                            f'Accuracy_{self.exp1_name}': f"{row1['Accuracy']:.2f}",
                            f'Accuracy_{self.exp2_name}': f"{row2['Accuracy']:.2f}",
                            'Accuracy_Œî': f"{(row2['Accuracy'] - row1['Accuracy']):.2f}",
                            f'Precision_{self.exp1_name}': f"{row1['Precision']:.2f}",
                            f'Precision_{self.exp2_name}': f"{row2['Precision']:.2f}",
                            'Precision_Œî': f"{(row2['Precision'] - row1['Precision']):.2f}",
                            f'F1_{self.exp1_name}': f"{row1['F1']:.2f}",
                            f'F1_{self.exp2_name}': f"{row2['F1']:.2f}",
                            'F1_Œî': f"{(row2['F1'] - row1['F1']):.2f}",
                            f'Jaccard_{self.exp1_name}': f"{row1['Jaccard']:.3f}",
                            f'Jaccard_{self.exp2_name}': f"{row2['Jaccard']:.3f}",
                            'Jaccard_Œî': f"{(row2['Jaccard'] - row1['Jaccard']):.3f}",
                        })
        
        comparison_df = pd.DataFrame(comparison_data)
        
        # Create table visualization
        headers = ['<b>Model</b>', '<b>Approach</b>', 
                  f'<b>Samples</b><br>{self.exp1_name}', f'<b>Samples</b><br>{self.exp2_name}',
                  f'<b>Accuracy</b><br>{self.exp1_name}', f'<b>Accuracy</b><br>{self.exp2_name}', '<b>Œî Acc</b>',
                  f'<b>Precision</b><br>{self.exp1_name}', f'<b>Precision</b><br>{self.exp2_name}', '<b>Œî Prec</b>',
                  f'<b>F1</b><br>{self.exp1_name}', f'<b>F1</b><br>{self.exp2_name}', '<b>Œî F1</b>',
                  f'<b>Jaccard</b><br>{self.exp1_name}', f'<b>Jaccard</b><br>{self.exp2_name}', '<b>Œî Jacc</b>']
        
        # Color code improvements (green for positive, red for negative)
        cell_colors = []
        for col in comparison_df.columns:
            if 'Œî' in col:
                colors = []
                for val in comparison_df[col]:
                    try:
                        num_val = float(val)
                        if num_val > 0:
                            colors.append('lightgreen')
                        elif num_val < 0:
                            colors.append('lightcoral')
                        else:
                            colors.append('white')
                    except:
                        colors.append('white')
                cell_colors.append(colors)
            else:
                cell_colors.append(['white'] * len(comparison_df))
        
        fig = go.Figure(data=[go.Table(
            header=dict(
                values=headers,
                fill_color='#2E86AB',
                align='center',
                font=dict(size=11, color='white', family='Times New Roman'),
                height=40
            ),
            cells=dict(
                values=[comparison_df[col] for col in comparison_df.columns],
                fill_color=cell_colors,
                align='center',
                font=dict(size=10, family='Times New Roman'),
                height=35
            )
        )])
        
        fig.update_layout(
            title=f"NL2aTOMIC Performance Comparison: {self.exp1_name} vs {self.exp2_name}",
            font=dict(family="Times New Roman", size=14),
            height=max(400, len(comparison_df) * 40 + 100),
            margin=dict(l=20, r=20, t=80, b=20)
        )
        
        return fig, comparison_df
    
    def create_metrics_comparison_chart(self):
        """Create comprehensive metrics comparison charts"""
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=('Accuracy Comparison', 'Precision vs Recall', 
                          'F1 Score vs Jaccard Index', 'Levenshtein Distance'),
            specs=[[{"secondary_y": False}, {"secondary_y": False}],
                   [{"secondary_y": False}, {"secondary_y": False}]]
        )
        
        # Accuracy comparison
        unique_combinations = self.df_combined[['Model', 'Approach']].drop_duplicates()
        
        for _, combo in unique_combinations.iterrows():
            model, approach = combo['Model'], combo['Approach']
            
            exp1_data = self.df1[(self.df1['Model'] == model) & (self.df1['Approach'] == approach)]
            exp2_data = self.df2[(self.df2['Model'] == model) & (self.df2['Approach'] == approach)]
            
            if len(exp1_data) > 0 and len(exp2_data) > 0:
                x_pos = f"{model}<br>{approach}"
                
                fig.add_trace(go.Bar(
                    name=self.exp1_name,
                    x=[x_pos],
                    y=[exp1_data['Accuracy'].iloc[0]],
                    marker_color=self.colors['primary'],
                    text=[f"{exp1_data['Accuracy'].iloc[0]:.1f}"],
                    textposition='auto',
                    showlegend=(_ == 0)
                ), row=1, col=1)
                
                fig.add_trace(go.Bar(
                    name=self.exp2_name,
                    x=[x_pos],
                    y=[exp2_data['Accuracy'].iloc[0]],
                    marker_color=self.colors['secondary'],
                    text=[f"{exp2_data['Accuracy'].iloc[0]:.1f}"],
                    textposition='auto',
                    showlegend=(_ == 0)
                ), row=1, col=1)
        
        # Precision vs Recall scatter
        fig.add_trace(go.Scatter(
            x=self.df1['Precision'],
            y=self.df1['Recall'],
            mode='markers+text',
            name=self.exp1_name,
            text=[f"{model}<br>{approach}" for model, approach in zip(self.df1['Model'], self.df1['Approach'])],
            textposition='top center',
            marker=dict(size=10, color=self.colors['primary']),
            showlegend=False
        ), row=1, col=2)
        
        fig.add_trace(go.Scatter(
            x=self.df2['Precision'],
            y=self.df2['Recall'],
            mode='markers+text',
            name=self.exp2_name,
            text=[f"{model}<br>{approach}" for model, approach in zip(self.df2['Model'], self.df2['Approach'])],
            textposition='bottom center',
            marker=dict(size=10, color=self.colors['secondary']),
            showlegend=False
        ), row=1, col=2)
        
        # F1 vs Jaccard scatter
        fig.add_trace(go.Scatter(
            x=self.df1['F1'],
            y=self.df1['Jaccard'],
            mode='markers+text',
            name=self.exp1_name,
            text=[f"{model}" for model in self.df1['Model']],
            textposition='top center',
            marker=dict(size=10, color=self.colors['primary']),
            showlegend=False
        ), row=2, col=1)
        
        fig.add_trace(go.Scatter(
            x=self.df2['F1'],
            y=self.df2['Jaccard'],
            mode='markers+text',
            name=self.exp2_name,
            text=[f"{model}" for model in self.df2['Model']],
            textposition='bottom center',
            marker=dict(size=10, color=self.colors['secondary']),
            showlegend=False
        ), row=2, col=1)
        
        # Levenshtein distance comparison
        for _, combo in unique_combinations.iterrows():
            model, approach = combo['Model'], combo['Approach']
            
            exp1_data = self.df1[(self.df1['Model'] == model) & (self.df1['Approach'] == approach)]
            exp2_data = self.df2[(self.df2['Model'] == model) & (self.df2['Approach'] == approach)]
            
            if len(exp1_data) > 0 and len(exp2_data) > 0:
                x_pos = f"{model}<br>{approach}"
                
                fig.add_trace(go.Bar(
                    name=self.exp1_name,
                    x=[x_pos],
                    y=[exp1_data['Levenshtein'].iloc[0]],
                    marker_color=self.colors['primary'],
                    text=[f"{exp1_data['Levenshtein'].iloc[0]:.2f}"],
                    textposition='auto',
                    showlegend=False
                ), row=2, col=2)
                
                fig.add_trace(go.Bar(
                    name=self.exp2_name,
                    x=[x_pos],
                    y=[exp2_data['Levenshtein'].iloc[0]],
                    marker_color=self.colors['secondary'],
                    text=[f"{exp2_data['Levenshtein'].iloc[0]:.2f}"],
                    textposition='auto',
                    showlegend=False
                ), row=2, col=2)
        
        fig.update_layout(
            title="NL2aTOMIC Comprehensive Metrics Analysis",
            font=dict(family="Times New Roman", size=12),
            height=800,
            barmode='group',
            legend=dict(
                orientation="h",
                yanchor="bottom",
                y=1.02,
                xanchor="right",
                x=1
            )
        )
        
        # Update axis labels
        fig.update_xaxes(title_text="Model-Approach", row=1, col=1)
        fig.update_yaxes(title_text="Accuracy (%)", row=1, col=1)
        fig.update_xaxes(title_text="Precision", row=1, col=2)
        fig.update_yaxes(title_text="Recall", row=1, col=2)
        fig.update_xaxes(title_text="F1 Score", row=2, col=1)
        fig.update_yaxes(title_text="Jaccard Index", row=2, col=1)
        fig.update_xaxes(title_text="Model-Approach", row=2, col=2)
        fig.update_yaxes(title_text="Levenshtein Distance", row=2, col=2)
        
        return fig
    
    def create_improvement_analysis(self):
        """Create detailed improvement analysis"""
        improvements = []
        
        for model in set(self.df1['Model']):
            for approach in set(self.df1[self.df1['Model'] == model]['Approach']):
                exp1_row = self.df1[(self.df1['Model'] == model) & (self.df1['Approach'] == approach)]
                exp2_row = self.df2[(self.df2['Model'] == model) & (self.df2['Approach'] == approach)]
                
                if len(exp1_row) > 0 and len(exp2_row) > 0:
                    exp1_row = exp1_row.iloc[0]
                    exp2_row = exp2_row.iloc[0]
                    
                    improvements.append({
                        'Model': model,
                        'Approach': approach,
                        'Accuracy_Œî': exp2_row['Accuracy'] - exp1_row['Accuracy'],
                        'Precision_Œî': exp2_row['Precision'] - exp1_row['Precision'],
                        'Recall_Œî': exp2_row['Recall'] - exp1_row['Recall'],
                        'F1_Œî': exp2_row['F1'] - exp1_row['F1'],
                        'Jaccard_Œî': exp2_row['Jaccard'] - exp1_row['Jaccard'],
                        'Levenshtein_Œî': exp2_row['Levenshtein'] - exp1_row['Levenshtein']
                    })
        
        imp_df = pd.DataFrame(improvements)
        
        # Create stacked bar chart for improvements
        fig = go.Figure()
        
        metrics = ['Accuracy_Œî', 'Precision_Œî', 'Recall_Œî', 'F1_Œî', 'Jaccard_Œî']
        colors = [self.colors['primary'], self.colors['secondary'], self.colors['accent1'], 
                 self.colors['accent2'], self.colors['accent3']]
        
        x_labels = [f"{row['Model']}<br>{row['Approach']}" for _, row in imp_df.iterrows()]
        
        for i, metric in enumerate(metrics):
            fig.add_trace(go.Bar(
                name=metric.replace('_Œî', ''),
                x=x_labels,
                y=imp_df[metric],
                marker_color=colors[i],
                text=[f'{val:.2f}' for val in imp_df[metric]],
                textposition='auto'
            ))
        
        fig.update_layout(
            title=f"Performance Improvements: {self.exp2_name} vs {self.exp1_name}",
            xaxis_title="Model-Approach Combinations",
            yaxis_title="Improvement (Œî)",
            font=dict(family="Times New Roman", size=12),
            height=600,
            barmode='group',
            legend=dict(
                orientation="h",
                yanchor="bottom",
                y=1.02,
                xanchor="right",
                x=1
            )
        )
        
        # Add horizontal line at y=0
        fig.add_hline(y=0, line_dash="dash", line_color="black", opacity=0.5)
        
        return fig, imp_df
    
    def create_radar_chart(self):
        """Create radar chart comparing performance across metrics"""
        # Average performance for each experiment
        metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'Jaccard']
        
        exp1_avg = [self.df1[metric].mean() for metric in metrics]
        exp2_avg = [self.df2[metric].mean() for metric in metrics]
        
        fig = go.Figure()
        
        fig.add_trace(go.Scatterpolar(
            r=exp1_avg,
            theta=metrics,
            fill='toself',
            name=self.exp1_name,
            marker_color=self.colors['primary']
        ))
        
        fig.add_trace(go.Scatterpolar(
            r=exp2_avg,
            theta=metrics,
            fill='toself',
            name=self.exp2_name,
            marker_color=self.colors['secondary']
        ))
        
        fig.update_layout(
            polar=dict(
                radialaxis=dict(
                    visible=True,
                    range=[0, max(max(exp1_avg), max(exp2_avg)) * 1.1]
                )),
            title="Average Performance Radar Chart",
            font=dict(family="Times New Roman", size=12),
            height=500
        )
        
        return fig
    
    def generate_statistical_summary(self):
        """Generate statistical summary and insights"""
        summary = {
            'datasets': {
                self.exp1_name: {
                    'n_models': len(self.df1['Model'].unique()),
                    'n_approaches': len(self.df1['Approach'].unique()),
                    'total_samples': self.df1['Total_Samples'].sum(),
                    'avg_accuracy': self.df1['Accuracy'].mean(),
                    'avg_precision': self.df1['Precision'].mean(),
                    'avg_recall': self.df1['Recall'].mean(),
                    'avg_f1': self.df1['F1'].mean(),
                    'avg_jaccard': self.df1['Jaccard'].mean(),
                    'avg_levenshtein': self.df1['Levenshtein'].mean()
                },
                self.exp2_name: {
                    'n_models': len(self.df2['Model'].unique()),
                    'n_approaches': len(self.df2['Approach'].unique()),
                    'total_samples': self.df2['Total_Samples'].sum(),
                    'avg_accuracy': self.df2['Accuracy'].mean(),
                    'avg_precision': self.df2['Precision'].mean(),
                    'avg_recall': self.df2['Recall'].mean(),
                    'avg_f1': self.df2['F1'].mean(),
                    'avg_jaccard': self.df2['Jaccard'].mean(),
                    'avg_levenshtein': self.df2['Levenshtein'].mean()
                }
            }
        }
        
        # Calculate improvements
        improvements = {
            'accuracy': summary['datasets'][self.exp2_name]['avg_accuracy'] - summary['datasets'][self.exp1_name]['avg_accuracy'],
            'precision': summary['datasets'][self.exp2_name]['avg_precision'] - summary['datasets'][self.exp1_name]['avg_precision'],
            'recall': summary['datasets'][self.exp2_name]['avg_recall'] - summary['datasets'][self.exp1_name]['avg_recall'],
            'f1': summary['datasets'][self.exp2_name]['avg_f1'] - summary['datasets'][self.exp1_name]['avg_f1'],
            'jaccard': summary['datasets'][self.exp2_name]['avg_jaccard'] - summary['datasets'][self.exp1_name]['avg_jaccard'],
            'levenshtein': summary['datasets'][self.exp2_name]['avg_levenshtein'] - summary['datasets'][self.exp1_name]['avg_levenshtein']
        }
        
        return summary, improvements
    
    def print_insights(self):
        """Print comprehensive insights"""
        summary, improvements = self.generate_statistical_summary()
        
        print("="*80)
        print("NL2aTOMIC PROPOSITION EXTRACTION - COMPARATIVE ANALYSIS")
        print("="*80)
        
        print(f"\nüìä DATASET OVERVIEW:")
        print(f"   {self.exp1_name}: {summary['datasets'][self.exp1_name]['n_models']} models, {summary['datasets'][self.exp1_name]['n_approaches']} approaches")
        print(f"   {self.exp2_name}: {summary['datasets'][self.exp2_name]['n_models']} models, {summary['datasets'][self.exp2_name]['n_approaches']} approaches")
        
        print(f"\nüéØ PERFORMANCE IMPROVEMENTS ({self.exp2_name} vs {self.exp1_name}):")
        print(f"   Accuracy:     {improvements['accuracy']:+.2f}%")
        print(f"   Precision:    {improvements['precision']:+.2f}%")
        print(f"   Recall:       {improvements['recall']:+.2f}%")
        print(f"   F1 Score:     {improvements['f1']:+.2f}%")
        print(f"   Jaccard:      {improvements['jaccard']:+.3f}")
        print(f"   Levenshtein:  {improvements['levenshtein']:+.2f}")
        
        print(f"\nüèÜ BEST PERFORMING MODELS:")
        best_f1_exp1 = self.df1.loc[self.df1['F1'].idxmax()]
        best_f1_exp2 = self.df2.loc[self.df2['F1'].idxmax()]
        print(f"   {self.exp1_name}: {best_f1_exp1['Model']} ({best_f1_exp1['Approach']}) - F1: {best_f1_exp1['F1']:.2f}")
        print(f"   {self.exp2_name}: {best_f1_exp2['Model']} ({best_f1_exp2['Approach']}) - F1: {best_f1_exp2['F1']:.2f}")
        
        print(f"\nüìà KEY INSIGHTS:")
        if improvements['accuracy'] > 0:
            print(f"   ‚úÖ {self.exp2_name} shows improved accuracy overall")
        else:
            print(f"   ‚ùå {self.exp1_name} performs better in accuracy")
            
        if improvements['levenshtein'] < 0:
            print(f"   ‚úÖ {self.exp2_name} produces text closer to ground truth (lower Levenshtein)")
        else:
            print(f"   ‚ùå {self.exp1_name} produces text closer to ground truth")
        
        print("\n" + "="*80)

# USAGE EXAMPLE:
# ===============

# Example data matching your structure
sample_data1 = {
    'Model': ['claude-sonnet', 'gpt-4', 'llama-2', 'mistral-7b'],
    'Approach': ['few_shot', 'few_shot', 'few_shot', 'few_shot'],
    'Total_Samples': [143, 143, 143, 143],
    'Accuracy': [52.19, 48.25, 45.80, 44.10],
    'Precision': [58.45, 55.20, 52.30, 50.15],
    'Recall': [59.64, 56.10, 53.20, 51.80],
    'F1': [57.96, 54.80, 52.10, 50.25],
    'Jaccard': [0.83, 0.78, 0.75, 0.72],
    'Levenshtein': [3.66, 4.12, 4.58, 4.85]
}

sample_data2 = {
    'Model': ['claude-sonnet', 'gpt-4', 'llama-2', 'mistral-7b'],
    'Approach': ['few_shot', 'few_shot', 'few_shot', 'few_shot'],
    'Total_Samples': [143, 143, 143, 143],
    'Accuracy': [58.74, 54.50, 51.20, 49.30],
    'Precision': [64.20, 60.85, 57.40, 55.25],
    'Recall': [65.10, 61.20, 58.10, 56.40],
    'F1': [63.45, 60.15, 56.80, 54.90],
    'Jaccard': [0.91, 0.86, 0.82, 0.79],
    'Levenshtein': [3.12, 3.58, 3.95, 4.22]
}

# Create DataFrames
# df1 = pd.DataFrame(sample_data1)
# df2 = pd.DataFrame(sample_data2)
df1 = pd.read_csv("/Users/priscilladanso/Documents/GitHub/prompt_dashboard/data/undefined/nl2pl_aggregated_results.csv")
df2 = pd.read_csv("/Users/priscilladanso/Documents/GitHub/prompt_dashboard/data/defined/nl2pl_aggregated_results.csv")

print("üìã SAMPLE DATA LOADED")
print("Replace 'sample_data1' and 'sample_data2' with your actual CSV data")
print("Use: df1 = pd.read_csv('your_dataset1.csv') and df2 = pd.read_csv('your_dataset2.csv')")
print("\n" + "="*60)

# Initialize analyzer
analyzer = NL2AtomicAnalyzer(df1, df2, "Minimal Prompt", "Detailed Prompt")

# Generate insights
analyzer.print_insights()

# Create visualizations
print("\nüé® GENERATING VISUALIZATIONS...")

# 1. Comprehensive comparison table
table_fig, comparison_df = analyzer.create_comprehensive_comparison_table()
table_fig.show()

# 2. Metrics comparison charts
metrics_fig = analyzer.create_metrics_comparison_chart()
metrics_fig.show()

# 3. Improvement analysis
improvement_fig, improvement_df = analyzer.create_improvement_analysis()
improvement_fig.show()

# 4. Radar chart
radar_fig = analyzer.create_radar_chart()
radar_fig.show()

print("\nüíæ EXPORT COMMANDS FOR PAPER:")
table_fig.write_image('nl2atomic_comparison_table.png', width=1400, height=600, scale=2)
print("table_fig.write_image('nl2atomic_comparison_table.png', width=1400, height=600, scale=2)")
metrics_fig.write_image('nl2atomic_metrics_analysis.png', width=1200, height=800, scale=2)
print("metrics_fig.write_image('nl2atomic_metrics_analysis.png', width=1200, height=800, scale=2)")
improvement_fig.write_image('nl2atomic_improvements.png', width=1200, height=600, scale=2)
print("improvement_fig.write_image('nl2atomic_improvements.png', width=1200, height=600, scale=2)")
print("radar_fig.write_image('nl2atomic_radar.png', width=800, height=500, scale=2)")
radar_fig.write_image('nl2atomic_radar.png', width=800, height=500, scale=2)
print("\n‚úÖ Analysis Complete! All visualizations generated.")
```
